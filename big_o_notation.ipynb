## Curves to graph

y = 0x + 1 and label the curve "O(1)"
y = log(x) and label the curve "O(log n)"
y = x and label the curve "O(n)"
y = x * log(x) and label it "O(n log n)"
y = x^2 and label it "O(n^2)"
y = 2^x and label it "O(2^n)"
y = n! and label is "O(n!)"
y = n^n and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(plt)
import math
```

```python
plt.figure(figsize = (16,9))

x1 = [x/10 for x in range(20)]
y1 = [(num * 0 + 1) for num in x1]
plt.plot(x1,y1, label = "O(1)")


x2 = [x/10 for x in range(10,20)]
y2 = [math.log(num) for num in x2]
plt.plot(x2,y2, label = "O(log n)")

x3 = [x/10 for x in range(20)]
y3 = [num for num in x3]
plt.plot(x3, y3, label = "O(n)")

x4 = [x/10 for x in range(10, 20)]
y4 = [(num * math.log(num)) for num in x4]
plt.plot(x4, y4, label = "O(n log n)")

x5 = [x/10 for x in range(20)]
y5 = [num ** 2 for num in x5]
plt.plot(x5,y5, label = "O(n^2)")

x6 = [x/10 for x in range(20)]
y6 = [2 ** num for num in x6]
plt.plot(x6,y6, label = "O(2^n)")

x7 = range(20)
y7 = [math.factorial(num) for num in x7]

x8 = [x/10 for x in range(20)]
y8 = [[num ** num] for num in x8]
plt.plot(x8,y8, label = "O(n^n)")


plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
```
